```
path_txt = "newsela_articles_20150302.aligned.sents.txt"
data_txt = pd.read_csv(path_txt, sep='\t', header=None, names=['doc', 'Vh', 'Vs', 'Complex', 'Simple'])
sentences = data_txt.dropna(subset=['Simple'])

# sentences = data_txt[['Vh', 'Vs', 'Complex', 'Simple']]
sentence_id = [0]
prev = [sentences.iloc[0]['Complex'], sentences.iloc[0]['Simple']]
for i in range(1, len(sentences)):
  if sentences.iloc[i]['Complex'] in prev:
    prev = list(prev)
    prev.extend([sentences.iloc[i]['Complex'], sentences.iloc[i]['Simple']])
    prev = set(prev)
    sentence_id.append(sentence_id[-1])
  else:
    sentence_id.append(sentence_id[-1] + 1)
    prev =  [sentences.iloc[i]['Complex'], sentences.iloc[i]['Simple']]

sentences['sentence_id'] = sentence_id
sentences
```

```
def find_simp(row):
  ver, sent_id = row['Version'], row['sent_id']
  if ver != 'V4':
    tmp_row = sentences['Simple'].loc[(sentences['Vh'] == ver) & (sentences['sentence_id'] == sent_id)]
    try:
      return tmp_row.iloc[0]
    except IndexError:
      return None
```

```
df['Simple Context'] = df.apply(lambda x: find_simp(x), axis=1)
```

```
def quest_exist(row):
  vers, s_id = row['Vh'], row['sentence_id']
  quests = df.loc[(df['Version']== vers) & (df['sent_id'] == s_id)]
  if len(quests) > 0: 
    return True
  else:
    vers = row['Vs']
    quests = df.loc[(df['Version']== vers) & (df['sent_id'] == s_id)]
    return True if len(quests)> 0 else False
```

```
sentences['Question'] = sentences.apply(lambda x: quest_exist(x), axis=1)
```

```
path = "/content/drive/My Drive/model_newsela/best_tfmr"
# path = "valhalla/t5-small-e2e-qg"
model = AutoModelForSeq2SeqLM.from_pretrained(path)
tokenizer = AutoTokenizer.from_pretrained(path)
model.eval()

def run_model(input_string, **generator_args):
    input_ids = tokenizer.encode(input_string, return_tensors="pt")
    res = model.generate(input_ids, **generator_args)
    tokens = [tokenizer.decode(x) for x in res]
    return(tokens)
```

## Calculate scores


```
final_result_df = pd.read_csv("/content/drive/MyDrive/final_result_df.csv")
```

```
from rouge_score import rouge_scorer,scoring 
import math

scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
aggregator = scoring.BootstrapAggregator()
sep = '<sep>'

def calculate_rouge(row):
  ref = row['Questions']
  gen_ques = row['Generated Question']
  if type(gen_ques) == str:
    scores = scorer.score(gen_ques, ref)
    return [round(v.fmeasure * 100, 4) for k, v in scores.items()][0]
  else:
    return 0

def calculate_sari(row):
  ref = row['Simple Context']
  context = row['Context']
  gen_simp = row['Generated Context']
  return SARIsent(context, gen_simp, ref)    
```

```
final_result_df['ROUGE score'] = final_result_df.apply(lambda x: calculate_rouge(x), axis=1)
# SARI?
final_result_df['SARI score'] = final_result_df.apply(lambda x: calculate_sari(x), axis=1)
```
